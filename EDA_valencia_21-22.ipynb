# Análisis de datos
### EDA con Pandas

```python
# instalar via pip en un venv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

calidad_aire = pd.read_csv("rvvcca_d_horarios_2021_2022.csv", sep=";")
calidad_aire_old = calidad_aire.copy()

# Mostrar las primeras filas del DataFrame
print("*** HEAD ***")
display(calidad_aire.head())
print("="*100)

# Mostrar la estructura del DataFrame
print("*** INFO ***")
display(calidad_aire.info())
print("="*100)
# Mostrar un resumen estadístico de las variables numéricas
print("*** DESC ***")
display(calidad_aire.describe())
print("="*100)
```

## Eliminamos datos innecesarios para el estudio

```python
calidad_aire = calidad_aire.drop(
    columns=[
        "Id",
        "Velocidad del viento",
        "Direccion del viento",
        "Ruido",
        "Temperatura",
        "Humedad relativa",
        "Presion",
        "Radiacion",
        "Precipitacion",
        "Velocidad maxima del viento",
        "Fecha creacion",
        "Fecha baja",
    ]
)
```

## Normalizamos datos

```python
# Normalizamos str
calidad_aire.columns = calidad_aire.columns.str.replace(" ", "_")

#COMBINANDO FECHA Y HORA
# Fecha
calidad_aire["Fecha"] = pd.to_datetime(
    calidad_aire["Fecha"],
    errors="coerce"
)

# Hora compatible con formatos mixtos
calidad_aire["Hora"] = pd.to_datetime(
    calidad_aire["Hora"].astype(str).str.strip(),
    format="mixed",
    errors="coerce"
).dt.strftime("%H:%M:%S")

# Combinar
calidad_aire["fecha_y_hora"] = pd.to_datetime(
    calidad_aire["Fecha"].astype(str) +
    " " +
    calidad_aire["Hora"],
    errors="coerce"
)

# Verificar
calidad_aire[["Fecha", "Hora", "fecha_y_hora"]].head()

# Estación a category
calidad_aire["Estacion"] = calidad_aire["Estacion"].astype("category")
# Dia de la semana a category
calidad_aire["Dia_de_la_semana"] = calidad_aire["Dia_de_la_semana"].astype("category")
```

## Mostramos valores faltantes por estación (gráficas)

```python
gases = ["CO", "NH3", "C7H8", "C6H6", "C8H10", "PM1"]

for gas in gases:
    
    # indicador temporal de nulos
    missing = calidad_aire[gas].isnull().astype(int)
    
    # proporción de nulos por estación
    resumen = (
        calidad_aire
        .groupby("Estacion")
        .apply(lambda x: x[gas].isnull().mean())
        .sort_values(ascending=False)
    )
    
    # mostrar tabla
    print(f"\n--- {gas} ---")
    display(resumen)
    
    # gráfico
    plt.figure(figsize=(10,4))
    
    resumen.plot(kind="bar")
    
    plt.ylabel("% valores faltantes")
    plt.title(f"Faltantes de {gas} por estación")
    
    plt.xticks(rotation=45)
    
    plt.show()
```

## Eliminamos estaciones meteorologicas

```python
calidad_aire = calidad_aire[
    ~calidad_aire["Estacion"].isin(["Nazaret Meteo", "Conselleria Meteo"])
]
calidad_aire["Estacion"].unique()
```

## Mostramos porcentaje de no nulos agrupados por estación (tabla)

```python
contaminantes = [
    "PM1", "PM2.5", "PM10",
    "NO", "NO2", "NOx",
    "O3", "SO2", "CO",
    "NH3", "C7H8", "C6H6",
    "C8H10"
]
cobertura = calidad_aire.groupby("Estacion")[contaminantes].apply(
    lambda x: x.notnull().mean()
)

display(cobertura)
```

## Cobertura de contaminantes por estación

```python
num_variables = (cobertura > 0.5).sum(axis=1)

num_variables = num_variables.sort_values(ascending=False)

display(num_variables)
num_variables.plot(
    kind="bar",
    figsize=(10,4)
)

plt.ylabel("Número de contaminantes medidos")
plt.title("Cobertura de contaminantes por estación")

plt.show()
```

## Concentracion media de NO2 por estacion

```python
media_no2 = calidad_aire.groupby("Estacion")["NO2"].mean()

media_no2 = media_no2.sort_values(ascending=False)

display(media_no2)
media_no2.plot(
    kind="bar",
    figsize=(10,4)
)

plt.ylabel("NO2 medio")
plt.title("Concentración media de NO2 por estación")

plt.show()
```

## Variabilidad entre estaciones (std)

```python
medias_estacion = calidad_aire.groupby("Estacion")[
    contaminantes
].mean()
variabilidad = medias_estacion.std()

variabilidad = variabilidad.sort_values(ascending=False)

display(variabilidad)
variabilidad.plot(
    kind="bar",
    figsize=(10,4)
)

plt.ylabel("Desviación típica")
plt.title("Variabilidad entre estaciones")

plt.show()
```

```python
cols_gases = ["CO", "NH3", "C7H8", "C6H6", "C8H10", "PM1"]
df_gases = calidad_aire[["fecha_y_hora", "Estacion"] + cols_gases].copy()
# ELIMINANDO COLUMNAS SOBRANTES
# Agregar columnas adicionales
cols_a_eliminar = list(["Fecha", "Hora"])
cols_a_eliminar.extend(cols_gases)

# Verificar
display(cols_a_eliminar)

# Eliminar columnas
calidad_aire = calidad_aire.drop(columns=cols_a_eliminar)
calidad_aire.info()
df_gases.info()
```

# Histogramas

```python
# Columnas numéricas (excluyendo irrelevantes)
excluir = ["Id", "Dia_del_mes"]

cols_numericas = calidad_aire.select_dtypes(
    include=["int64", "float64"]
).columns

cols_numericas = [c for c in cols_numericas if c not in excluir]

# Definir grid
n_cols = 4  # columnas del grid
n_rows = int(np.ceil(len(cols_numericas) / n_cols))

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 3 * n_rows))

axes = axes.flatten()

# Dibujar histogramas
for i, col in enumerate(cols_numericas):
    axes[i].hist(calidad_aire[col].dropna(), bins=30)
    axes[i].set_title(col)
    axes[i].tick_params(axis='x', rotation=45)

# Eliminar ejes vacíos
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

# TRATA DE DATOS AUSENTES

```python
calidad_aire = calidad_aire.ffill()
calidad_aire.info()
```

## Rango de fechas del dataset y horas faltantes por estacion

```python
rango = (
    calidad_aire["fecha_y_hora"].max() -
    calidad_aire["fecha_y_hora"].min()
)
print("Fecha mínima:", calidad_aire["fecha_y_hora"].min())
print("Fecha máxima:", calidad_aire["fecha_y_hora"].max())
print("Duración del dataset:", rango)


fecha_min = calidad_aire["fecha_y_hora"].min()
fecha_max = calidad_aire["fecha_y_hora"].max()

horas_completas = pd.date_range(
    start=fecha_min,
    end=fecha_max,
    freq="h"
)

for estacion in calidad_aire["Estacion"].unique():
    
    # datos de la estación
    df_estacion = calidad_aire[
        calidad_aire["Estacion"] == estacion
    ]
    
    # horas observadas
    horas_observadas = df_estacion["fecha_y_hora"]
    
    # horas faltantes
    horas_faltantes = horas_completas.difference(horas_observadas)
    
    print(f"\n--- {estacion} ---")
    print("Horas faltantes:", len(horas_faltantes))
```

## Cajas y bigotes

```python
exclude_cols = ['Dia_de_la_semana', 'Dia_del_mes']

num_cols = calidad_aire.select_dtypes(include=['int64', 'float64']).columns
num_cols = [col for col in num_cols if col not in exclude_cols]
for col in num_cols:
    Q1 = calidad_aire[col].quantile(0.25)
    Q3 = calidad_aire[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = calidad_aire[(calidad_aire[col] < lower) | (calidad_aire[col] > upper)]

    print(f"{col}: {len(outliers)} outliers")

    plt.figure(figsize=(6,4))
    sns.boxplot(x=calidad_aire[col])
    plt.xscale('log')
    plt.title(f'{col} (outliers: {len(outliers)})')
    plt.show()
```

```python
limites = {

    # Partículas
    "PM2.5": (0, 100),
    "PM10": (0, 175),

    # Gases
    "NO": (0, 150),
    "NO2": (0, 100),
    "NOx": (0, 150),
    "O3": (0, 50),
    "SO2": (0, 100),
}

#Remplazamos por NaN los valores fisicamente imposibles
for columna, (minimo, maximo) in limites.items():

    if columna in calidad_aire.columns:

        calidad_aire.loc[
            (calidad_aire[columna] < minimo) |
            (calidad_aire[columna] > maximo),
            columna
        ] = np.nan

#Ordenamos por fecha
calidad_aire = calidad_aire.sort_values(
    ["Estacion", "fecha_y_hora"]
)
display(
    (calidad_aire.isnull().mean() * 100)
)
calidad_aire.describe()
```
